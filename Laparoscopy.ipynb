```python
from ultralytics import YOLO

model = YOLO(r"C:\Users\sadiq\Cholec80-Surgical-Tool-Detection-main\models\best.pt")  # e.g. r"C:\Users\sadiq\lap_tools_run\yolov8s_laptools\weights\best.pt"

import cv2
from ultralytics import YOLO

VIDEO_PATH = r"C:\Users\sadiq\Lapchole1.mp4"
MODEL_PATH = r"C:\Users\sadiq\Cholec80-Surgical-Tool-Detection-main\models\best.pt"

model = YOLO(MODEL_PATH)

cap = cv2.VideoCapture(VIDEO_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
if fps <= 0:
    fps = 30
delay = int(1000 / fps)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, imgsz=640, conf=0.4)[0]

    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        cls = int(box.cls[0])
        conf = float(box.conf[0])
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(
            frame, f"{cls} {conf:.2f}", (int(x1), int(y1) - 5),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1
        )

    cv2.imshow("Lap Tools YOLO (best.pt)", frame)
    if cv2.waitKey(delay) & 0xFF in (27, ord("q")):
        break

cap.release()
cv2.destroyAllWindows()
```
